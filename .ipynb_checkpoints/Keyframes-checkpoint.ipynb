```python
import cv2.cv as cv
import numpy as np
import cv2
import sys
name = "resources/HansRosling.mp4"

def shots(t1, t2):
    cap = cv2.VideoCapture(name)
    cap.isOpened()
    
    fps = int(cap.get(cv2.cv.CV_CAP_PROP_FPS))
    t1_frame = t1*fps
    t2_frame = t2*fps
    
    minutes = (float(t2_frame - t1_frame)/float(fps)) / 60.0
    m1 = (float(t1_frame)/float(fps)) / 60.0
    m2 = (float(t2_frame)/float(fps)) / 60.0
    
    print "Processing ", minutes, " minutes of video"
    print "From minute ", m1, " to minute ", m2
    print "FPS: ", fps
    print "Total frames: ", t2_frame - t1_frame
    print "Processed frames: ", float(t2_frame - t1_frame)/float(fps)
    
    i = 0
    frames = []
    while i < t2_frame: 
        if i%fps == 0 and i > t1_frame:
            frames.append(frame)
        ret, frame = cap.read()
        i+=1
    print "Frame Length: ", len(frames)
    
    h = np.zeros((300,256,3))
    bins = np.arange(256).reshape(256,1)
    color = [ (255,0,0),(0,255,0),(0,0,255) ]
    ch = 0
    col = color[0]
    hists = []
    for frame in frames: 
        img = frame
        hist_item = cv2.calcHist([img],[ch],None,[256],[0,255])
        cv2.normalize(hist_item,hist_item,0,255,cv2.NORM_MINMAX)
        hists.append(np.array([i[0] for i in hist_item]))
    
    import scipy.spatial as spat
    S = 1 - spat.distance.squareform(spat.distance.pdist(np.array(hists), 'cosine'))
    
    return {'similarity': S, 'frames': frames}
```
