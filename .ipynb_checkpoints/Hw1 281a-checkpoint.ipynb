```python
# for pr(a | b) call
# pr_mat[a, b]
import copy
import numpy as np
pr_mat = np.array([[.6, .2],[.4, .8]])
possibles = np.array([0,1])

pr_x1 = .5
n = 4
graph = [n-i for i in range(n)]

al2 = []
final_prob = []

class Potential(object):
    def __init__(self):
        pass

class P(Potential):
    def __init__(self, node, parent, value):
        self.node = node
        self.parent = parent
        self.value = value
        self.nr = [node, parent]
    
    def val(self, node_poss, parent_poss):
        if node_poss is not None:
            self.value = self.value[node_poss, :]
        if parent_poss is not None:
            self.value = self.value[:, parent_poss]
            
        return self.value

class Delta(Potential):
    def __init__(self, node, value):
        self.node = node
        self.value = value
        self.arrval = [0, 0]
        self.arrval[self.value] = 1
        self.nr = [node]
    
    def val(self, poss):
        if not poss:
            self.value = self.arrval[:]
            return self.arrval
        if poss == self.value:
            self.value = 1
            return 1
        else:
            self.value = 0
            return 0

class M(Potential):
    def __init__(self, nodes, evalarr):
        self.nodes = nodes[:]
        self.evalarr = evalarr
        self.nr = self.nodes[:]
    
    def evaluate(self, earr):
        if type(earr) is P or type(earr) is Delta:
            return earr.value
        if type(earr) is M:
            return earr.val()
        if earr[0] is "sum":
            running = 0
            for x in earr[1]:
                running += self.evaluate(x)
            return running
        elif earr[0] is "prod":
            running = 1
            for x in earr[1]:
                running *= self.evaluate(x)
            return running
    
    def val(self):
        return self.evaluate(self.evalarr)
        

def m(potentials,i,target_node,target_val):
    eval_dict = ["sum", []]
    for possible in possibles:
        prod = []
        for potential in potentials:
            pot = copy.deepcopy(potential)
            if type(pot) is P:
                if pot.parent is i:
                    pot.val(None, possible)
                elif pot.node is i:
                    pot.val(possible, None)
            elif type(pot) is Delta:
                pot.val(possible)
            else:
                pot.val()
            prod.append(pot)

        if possible == target_val and i == target_node:
            ed2 = ["prod", prod]
            final_prob.append(M([], ed2).val())
        eval_dict[1].append(["prod", prod])
    nr = set()
    for potential in potentials:
        if type(potential) is not M:
            nr.add(potential.node)
        if type(potential) is P:
            nr.add(potential.parent)
        if type(potential) is M:
            for node in potential.nr:
                nr.add(node)
    if i in nr:
        nr.remove(i)
    nr = list(nr)
        
    return M(nr, eval_dict)
    
    
    
def init(g,f): 
    for node in g:
        al2.append(P(node, node - 1, pr_mat))

def evid(e, val): 
    al2.append(Delta(e, val))

def update(g, target_node, target_val): 
    for node in g: 
        pot = []
        ind = []
        
        for i in range(len(al2)):
            if node in al2[i].nr:
                pot.append(al2[i])
                ind.append(i)

        del al2[:]
        for i, a in enumerate(al2):
            if i not in ind:
                al2.append(a)

        m_pot = m(pot, node, target_node, target_val)
        al2.append(m_pot)

def normalize(f):
    norm_factor = al2[0].val()
    return final_prob[0] / float(norm_factor)
            
def eliminate(g, e, f):
    init(g, f)
    evid(e, 1)
    update(g, 1, 0)
    return normalize(f)
    
print eliminate(graph, 4, 1)
```
